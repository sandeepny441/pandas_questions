```python
df = pd.read_csv('df_apple.csv')
df
```

```python
df = df.sort_values(by = 'Price', ascending = False)
df['min_rank'] = df['Price'].rank(method = 'min', ascending = False)
df
```

```python
df['max_rank'] = df['Price'].rank(method = 'max', ascending = False)
df
```

```python
df['avg_rank'] = df['Price'].rank(method = 'average', ascending = False)
df
```

```python
df['pct_RANK'] = df['Price'].rank(pct = True, ascending = False)
df
```
