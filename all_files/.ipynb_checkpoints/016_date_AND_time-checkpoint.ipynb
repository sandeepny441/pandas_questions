```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Create a DataFrame with date and time columns
np.random.seed(0)
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)
date_list = [start_date + timedelta(days=random.randint(0, 364)) for _ in range(10)]
time_list = [datetime(2022, 1, 1) + timedelta(minutes=random.randint(0, 1439)) for _ in range(10)]

df_datetime = pd.DataFrame({
    'Date': date_list,
    'Time': time_list,
    'Value1': np.random.randint(1, 100, 10),
    'Value2': np.random.randint(100, 200, 10)
})

df_datetime
```
