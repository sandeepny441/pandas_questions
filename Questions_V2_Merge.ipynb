```python
import numpy as np
import pandas as pd


data = {
'Shop_Name': ['Starbucks', 'Blue Bottle', 'Verve Coffee', 'Stumptown', 'Blue Bottle', 'Gregorys Coffee', 'Cafe Grumpy'],
'Location': ['Hollywood', 'San Diego', 'Hollywood', 'San Diego', 'New York', 'New York', 'New York'],
'Rating': [4.5, 4.8, 4.4, 4.3, 4.6, 4.2, 4.7],
'Revenue': [50000, 30000, 25000, 28000, 52000, 20000, 24000]
}

coffee_shops = pd.DataFrame(data)
coffee_shops
```

```python
suppliers_data = {
'Shop_Name': ['Starbucks', 'Blue Bottle', 'Verve Coffee', 'Stumptown', 'Gregorys Coffee'],
'Supplier_Name': ['Beans R Us', 'Premium Beans', 'South Coffee Suppliers', 'San Diego Beans', 'NY Fresh Beans'],
'Delivery_Days': [7, 5, 7, 4, 6]
}

suppliers_df = pd.DataFrame(suppliers_data)

suppliers_df
```

```python
df_merged = pd.merge(coffee_shops, suppliers_df, on = 'Shop_Name')
df_merged
```

```python
df_merged = pd.merge(coffee_shops, suppliers_df, on = 'Shop_Name', how = "inner")
df_merged
```

```python
df_merged = pd.merge(coffee_shops, suppliers_df, on = 'Shop_Name', how = "outer")
df_merged
```

```python
df_merged = pd.merge(coffee_shops, suppliers_df, on = 'Shop_Name', how = "left")
df_merged
```

```python
df_merged = pd.merge(coffee_shops, suppliers_df, on = 'Shop_Name', how = "right")
df_merged
```

```python
df_merged = pd.merge(coffee_shops, suppliers_df, on = 'Shop_Name', how = "inner", suffixes = ('_df1', '_df2'))
df_merged
```
